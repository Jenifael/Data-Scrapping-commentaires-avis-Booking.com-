## Configurer le nombre d'avis que vous souhaitez télécharger

- Le nombre d'avis est par défaut mis sur 5*25=125 avis. (end_page) dans "properties" ci-dessous. Vous devez choisir le nombre de page que vous souhaitez que notre script étudie. Si vous souhaitez savoir à quoi cela correspond, allez sur le lien que vous avez modifié ci-dessus:

- Ici il y en a 50 max donc vous pouvez indiquer 50 à la variable "end_page" ci-dessous

```python
properties= {
#Page contenant les avis
  "base_page_url": "https://www.booking.com/reviewlist.fr.html?aid=304142;label=gen173nr-1FCAEoggI46AdIM1gEaE2IAQGYAQ24ARfIAQ_YAQHoAQH4AQuIAgGoAgO4AriJu_MFwAIB;sid=60a231f63268bc9ef8f65c0ac3f0d732;cc1=be;dist=1;pagename=housing;type=total&;rows=25",
  "next_page_url": ";offset=",
#Nombre de page que vous souhaitez
  "end_page": 5,
    
#Sleep time entre chaque connexion à booking.com 
  "minimum_sleep_time": 0,
  "maximum_sleep_time": 0,
#Versions des headers
  "minimum_major_version": 1,
  "maximum_major_version": 56,
  "minimum_minor_version": 1,
  "maximum_minor_version": 10,
}
```

# Script 

```python
# -*- coding: utf8 -*-
import time
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

#Beautiful soup
from bs4 import BeautifulSoup, element
import urllib
from random import randint, choice
import time
import os

os.chdir(r'C:\Users\Alex\Desktop\Mémoire') #Indiquez votre répertoire de travail ici
```

## Headers & Connexions 


Certains sites webs renvoient une erreur HTTP 429 lorsqu'on envoie trop de requêtes sur le site. Il existe un astuce pour contourner cela qui consiste à créer un utilisateur random.

```python
def create_random_header():

    browsers = ["Mozilla", "Chrome"]
    os_list = ["Windows NT 6.1; Win64; x64", "X11; Linux x86_64"]
    major_version = randint(properties['minimum_major_version'], properties['maximum_major_version'])
    minor_version = randint(properties['minimum_minor_version'], properties['maximum_minor_version'])
    chosen_browser = choice(browsers)
    chosen_os = choice(os_list)

    user_agent = '{}/{}.{} ({})'.format(
        chosen_browser,
        major_version,
        minor_version,
        chosen_os)
    header = {'User-Agent': user_agent}
    return header


def get_page(url):
    """
    Perform a GET request to the given URL and return results.
    Add a wait logic that, combined with random header, will help avoiding
    """
    header =create_random_header()
    request = urllib.request.Request(url, headers=header)
    result = urllib.request.urlopen(request).read()
    result=result.decode("utf-8")
    time.sleep(randint(properties['minimum_sleep_time'], properties['maximum_sleep_time']))
    return result,header
```

## Main script

```python
def main_script():
    """Booking.com est étrange, il renvoie des pages HTML aléatoires, même en gardant le même header.""" 
    """On va écrire une boucle qui va tester les url jusqu'à obtenir la page que l'on veut."""
    
    list_title=[]
    list_grade=[]
    list_namereviewer=[]
    list_countryreviewer=[]
    list_bedroomtype=[]
    list_commentadded=[]
    list_nightmonth=[]
    list_nightnb=[]
    list_reviewcommentliked=[]
    list_reviewcommentdisliked=[]
    
    for page in range(0,properties['end_page']):
        print('--->Reviewing page number : ',page+1)
        verif=-1
        verifsec=-1
        k=0
        while (verif<0) or (verifsec<0):
            page_url = "{}{}{}".format(properties['base_page_url'],properties['next_page_url'], str(page*25))
            current_page = get_page(page_url)[0]
            header=get_page(page_url)[1]
            verif=current_page.find('nuit')
            verifsec=current_page.find('bui-grid__column-9')
            k+=1
            print('--Tentative de connexion à la page numéro : {} \n Page : {} '.format(k,page_url))
        soup = BeautifulSoup(current_page, 'lxml')
        for review in soup.findAll("div", {"class": "bui-grid__column-9"}):

            #TITLE
            tofind=review.find("h3", {"class": "c-review-block__title"})
            list_title.append(tofind.get_text(strip=True)) if tofind != None else list_title.append("")

            #GRADE
            tofind=review.find("div", {"class": "bui-review-score__badge"})
            list_grade.append(tofind.get_text(strip=True)) if tofind != None else list_grade.append("")

            #NAME REVIEWER
            tofind=review.parent.find("span", {"class": "bui-avatar-block__title"})
            list_namereviewer.append(tofind.get_text(strip=True)) if tofind != None else list_namereviewer.append("")

            #COUNTRY REVIEWER
            tofind=review.parent.find("span", {"class": "bui-avatar-block__subtitle"})
            list_countryreviewer.append(tofind.get_text(strip=True)) if tofind != None else list_countryreviewer.append("")

            if len(review.parent.findAll("span", {"class": "c-review-block__date"}))==2:
                #DATE OF COMMENT
                list_commentadded.append(review.parent.findAll("span", {"class": "c-review-block__date"})[1].get_text(strip=True).split('le ')[1])
                #MONTH OF TRIP
                list_nightmonth.append(review.parent.findAll("span", {"class": "c-review-block__date"})[0].get_text(strip=True))
            else:
                list_commentadded.append("")
                list_nightmonth.append("")

            if len(review.parent.findAll("div", {"class": "bui-list__body"}))==2:
                #BEDROOM TYPE
                list_bedroomtype.append(review.parent.findAll("div", {"class": "bui-list__body"})[0].get_text(strip=True)) 
                #NUMBER OF NIGHT
                tofind=review.parent.findAll("div", {"class": "bui-list__body"})[1]
                for span in tofind.find_all('span'):
                        span.decompose()
                list_nightnb.append(tofind.get_text(strip=True)[0:2]) if tofind != None else list_nightnb.append("")
            else:
                list_nightnb.append("")
                list_bedroomtype.append("")

            #REVIEW A AIME
            revpos=review.find(text="A aimé")
            if revpos != None:
                revposfr=revpos.findNext('span').findNext('span').get_text(strip=True)
                if revposfr!='':
                    list_reviewcommentliked.append(revposfr)
                else:
                    list_reviewcommentliked.append(revpos.findNext('span').get_text(strip=True))
            else: list_reviewcommentliked.append("")

            #REVIEW N'A PAS AIME
            revneg=review.find(text="N'a pas aimé")
            if revneg != None:
                revnegfr=revneg.findNext('span').findNext('span').get_text(strip=True)
                if revnegfr!='':
                    list_reviewcommentdisliked.append(revnegfr)
                else:
                    list_reviewcommentdisliked.append(revneg.findNext('span').get_text(strip=True))
            else: list_reviewcommentdisliked.append("")

        print('Page {} done'.format(page+1))

    columns = {
            'Review Title': list_title,
            'Grade': list_grade,
            'Reviewer name': list_namereviewer,
            'Date review': list_commentadded,
            'Month of trip': list_nightmonth,
            'Number of nights':list_nightnb,
            'Country reviewer' : list_countryreviewer,
            'Bedroom Type': list_bedroomtype,
            'The reviewer liked': list_reviewcommentliked,
            'The reviewer did not liked': list_reviewcommentdisliked,
        }
    data_download=df = pd.DataFrame(columns)
    
    return data_download
```

```python
if __name__ == "__main__":

    
    df_review_Housingbrussels =main_script()
    df_review_Housingbrussels.to_csv('df_review_Housingbrussels.csv',sep=',',index=False,encoding='utf-8-sig')
```
